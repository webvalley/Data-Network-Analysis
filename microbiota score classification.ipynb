```python
import pandas as pd
import os
import math
```

```python
dataDir='./Data/'
filename='microbiota_clean_wSCORE.xlsx'

path = os.path.join(dataDir,filename)
```

```python
df = pd.read_excel(path)
```

```python
df.shape
```

```python
df = df[df.SCORE.notnull()]
```

```python
df.shape
```

```python
df['ScoreClass']=1
```

```python
df[['SCORE','ScoreClass']].head(300)
```

```python
tresholds=(1,2,5,100)
```

```python
for i, row in df.iterrows():
    for index, el in enumerate(tresholds):
        if (row['SCORE']>el and row['SCORE']<tresholds[index+1]):
            df['ScoreClass'][i]=index+2
            break
```
